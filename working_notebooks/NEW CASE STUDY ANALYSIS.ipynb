```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

import pickle
import os
os.chdir('C:/Users/renan/Desktop/full_count_2010')

import warnings
warnings.filterwarnings("ignore")
```

```python
with open('sys_application_result_notebook.pkl', 'rb') as input:
    sys = pickle.load(input)
    
with open('per_application_result_notebook.pkl', 'rb') as input:
    per = pickle.load(input)
    
with open('la_la_application_result_notebook.pkl', 'rb') as input:
    la_la = pickle.load(input)
    
with open('la_ny_application_result_notebook.pkl', 'rb') as input:
    la_ny = pickle.load(input)
```

# Los Angeles Systematic

```python
# Systematic Approach

# The iterator of the subplots is not zero index-based

fig = plt.figure(figsize=(30, 20))
fig.subplots_adjust(hspace = 0.35, wspace = 0.35) # hspace space between lines

for i in range(1, len(sys) + 1):
    
    aux = sys[i-1]
    
    ax = fig.add_subplot(5, 5, i)
    sns.distplot(aux.est_sim, 
                 color = 'darkblue', 
                 hist_kws={'edgecolor':'black'},
                 kde_kws={'linewidth': 2})
    plt.axvline(aux.statistic, color = 'red', linewidth = 2)
    plt.title('{} = {}'.format(aux._class_name, round(aux.statistic, 3)), fontweight = "bold", fontsize = 15)
```
