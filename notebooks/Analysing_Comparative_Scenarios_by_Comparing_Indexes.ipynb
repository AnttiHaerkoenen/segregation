```python
import pickle
import os
os.chdir('/bigdata/reylab/renancortes/ridir_related/')

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load results
with open('res_data.pkl', 'rb') as input:
    x = pickle.load(input)

# To vary the approach it is needed to vary x[i]
```

# First approach of analysing: plotting (to compare indexes)

```python
fig = plt.figure(figsize = (30, 20))
plt.suptitle('Increasing Global Population: Random Data', size = 50)

fig.subplots_adjust(hspace = 0.35, wspace = 0.35, top = 0.88) # hspace space between lines

for i in range(1, len(x[0])+1):
    
    aux = x[0][i-1][0]
    
    ax = fig.add_subplot(5, 5, i)
    sns.distplot(aux.est_sim, 
                 color = 'darkblue', 
                 hist_kws = {'edgecolor':'black'},
                 kde_kws = {'linewidth': 2})
    plt.axvline(aux.est_point_diff, color = 'red', linewidth = 2)
    plt.title('Dif. of {} = {}'.format(aux._class_name, round(aux.est_point_diff, 3)), fontweight = "bold", fontsize = 13)
```
