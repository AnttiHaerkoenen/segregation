```python
import pickle
import os
os.chdir('/bigdata/reylab/renancortes/ridir_related/')

import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
# Load results
with open('res_data.pkl', 'rb') as input:
    x = pickle.load(input)

# To vary the approach it is needed to vary x[i]
```

# Second approach of analysing: plotting (to compare approaches)

# Scenario 1

```python
fig = plt.figure(figsize = (14, 35))
plt.suptitle('Increasing Global Population', size = 35)

fig.subplots_adjust(hspace = 1.25, wspace = 0.2, top = 0.95) # hspace space between lines

aux_RD = x[0]
aux_PC = x[1]

for i in range(1, (len(x[0]) * 2) + 1):
    
    j = math.ceil(i/2)
    
    if (i % 2 == 1):
        aux = aux_RD[j-1][0]
        approach = "RD"
    else:
        aux = aux_PC[j-1][0]
        approach = "PC"
        
    ax = fig.add_subplot(23, 2, i)
    sns.distplot(aux.est_sim, 
                 color = 'darkblue', 
                 hist_kws = {'edgecolor':'black'},
                 kde_kws = {'linewidth': 2})
    plt.axvline(aux.est_point_diff, color = 'red', linewidth = 2)
    plt.title('Dif. of {} = {} ({})'.format(aux._class_name, 
                                            round(aux.est_point_diff, 3), 
                                            approach), 
              fontweight = "bold", 
              fontsize = 13)
plt.tight_layout(rect = [0, 0.03, 1, 0.95]) # To rearrange the subplots
```
